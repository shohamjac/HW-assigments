```python
%matplotlib inline
def plot_width(exp_list, exp_names,mul_list, hist_ran = (1e-8,5e-8)):
    print(hist_ran)
    time_list = []
    filtered_exc_list = []
    for multi in multiplier_list:
        fig, axes = plt.subplots(1, 1, figsize=(8, 4))
        for (i, exp) in enumerate(experiment_list):
            d = {"id": range(exp.number_of_detections),
                 "seconds": exp.get_time_list(),
                 "energy": exp.get_energy_list(),
                 "width": exp.get_width_list_based_on_peak(peak_multiplier=multi),
                 }
            exc_data = pd.DataFrame(data=d)
            #exc_data_filtered = exc_data[(exc_data.energy>energy_cut)]  # & (exc_data.steepness40_ratio>steepness_value)]
            axes.hist(exc_data.width, bins=100, histtype='step', range=hist_ran)
            axes.title.set_text(experiment_names[i])
            axes.set_ylabel("count")
            axes.set_xlim(list(hist_ran))
            plt.xlabel(f"width in seconds, based on {multi} of the peak")
            time_list.append(float(exc_data.seconds[-1:]-exc_data.seconds[0]))
            filtered_exc_list.append(len(exc_data))
        axes.legend(experiment_names)
        fig.tight_layout()
        fig.savefig("out.png", dpi=100)
        plt.show()
```

```python
experiment_list = [roi_ng38, roi_cs2]
experiment_names = ["ND3_ng38" , "ND3_cs"]
multiplier_list = [0.05, 0.1, 0.2]
plot_width(experiment_list, experiment_names, multiplier_list)
```
